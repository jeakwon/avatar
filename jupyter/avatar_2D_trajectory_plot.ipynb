## Import lib and check version

```python
import avatarpy
from avatarpy import Avatar, dataset
print(avatarpy.__version__)
```

## Avatar with sample dataset

```python
# check available dataset
for k, v in dataset.items():
    print(k, v)
```

```python
csv_path = dataset['freely_moving']
avatar = Avatar(csv_path, frame_rate=20) # frame_rate default is 20. frame_rate = recording speed of acquisition camera
```

## Plot 2D trajectory with chest node
Avatar instance has default 9 nodes. Among these I'll use `chest` node to plot 2D trajectory.  

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Available node list
avatar.nodes.keys()
```

```python
# Format of data
avatar['chest'].head()
```

```python
sns.set('talk')
sns.set_style('white')
g = sns.jointplot(data=avatar['chest'], x='x', y='y', kind='scatter', marker="+", s=10, color='orange')
g.plot_joint(sns.kdeplot, cmap='magma', zorder=0, thresh=0, levels=100, fill=True)
```
